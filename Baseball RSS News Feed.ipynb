```python
import pandas as pd
import feedparser as fp
```

The get_Headlines funcstion will pull the RSS feed from Google News and return a DataFrame of the players name, the

```python
def get_Headlines(player):
    link = 'https://news.google.com/rss/search?q='
    p_format = player.replace(' ','+').lower()
    search = link + p_format
    feed = fp.parse(search)
    df = pd.DataFrame()
    i = 0
    while i < 10:
        entry = feed.entries[i]
        temp = [[player,entry.title,entry.link]]
        df_temp = pd.DataFrame(temp)
        df = pd.concat([df,df_temp])
        i += 1
    df.columns = ['Player','Headline','URL']
    df = df.reset_index(drop = True)
    return df
```

```python
def main():
    players = ['Bryce Harper','Andrew Mccutchen','Mike Trout']
    output = pd.DataFrame()
    for player in players:
        temp = get_Headlines(player)
        output = pd.concat([output, temp])
    output = output.reset_index(drop = True)
    return output
```

```python
main()
```
